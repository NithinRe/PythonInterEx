```python
#Write a Python function that takes a sequence of numbers and determines whether all the numbers are different from each other
def test_data(data):
    if len(data) == len(set(data)):
        return True
    else:
        return False
print(test_data([1,2,3,4,5,6]))
```

```python
#Write a Python program to check the sum of three elements (each from an array) from three arrays is equal to a target value. Print all those three-element combinations.
import itertools
from functools import partial
X = [10,20,20,20]
Y = [10,20,30,40]
Z = [10,30,20,40]
T = 70

def check_sum_array(N,*num):
    if sum(x for x in num) == N:
        return (True,num)
    else:
        return (False,num)
pro = itertools.product(X,Y,Z)
func = partial(check_sum_array,T)
sums = list(itertools.starmap(func,pro))
result = set()
for s in sums:
    if s[0] == True and s[1] not in result:
        result.add(s[1])
        print(result)
```

```python
def permutation(num):
    result = [[]]
    for n in num:
        new_permutation = []
        for perm in result:
            for i in range(len(perm)+1):
                new_permutation.append(perm[:i] + [n] + perm[i:])
                result = new_permutation
        return result
nums = [1,2,3,4,5]
print('Orginal permutation : {}'.format(nums))
print('Collection of diffentent number {}'.format(permutation(nums)))

    
```

```python
def permute(nums):
  result_perms = [[]]
  for n in nums:
    new_perms = []
    for perm in result_perms:
      for i in range(len(perm)+1):
        new_perms.append(perm[:i] + [n] + perm[i:])
        result_perms = new_perms
  return result_perms

my_nums = [1,2,3,4,5]
print("Original Cofllection: ",my_nums)
print("Collection of distinct numbers:\n",permute(my_nums))
```
